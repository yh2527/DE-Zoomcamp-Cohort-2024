```python
import pandas as pd
import os
```

```python
# First, download the file using wget
url = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/'
color = 'green'
year = '2019'
month = '09'

exit_code = os.system(f"wget {url}{color}/{color}_tripdata_{year}-{month}.csv.gz")

if exit_code == 0:
    print("Download successful.")
else:
    print("Download failed. Exit code:", exit_code)
```

```python
# Reread the data into a iterator because we would like to ingest the file in smaller chunks
df_iter = pd.read_csv('green_tripdata_2019-09.csv.gz', iterator=True, chunksize=100000)
```

```python
df0 = next(df_iter)
len(df0)
```

```python
# get the column schema
df0.lpep_pickup_datetime = pd.to_datetime(df0.lpep_pickup_datetime)
df0.lpep_dropoff_datetime = pd.to_datetime(df0.lpep_dropoff_datetime)

df0.head(n=0).to_sql(name=f'{color}_tripdata_{year}_{month}', con=engine, if_exists='replace')
```

```python
query = """
SELECT *
FROM pg_catalog.pg_tables
WHERE schemaname != 'pg_catalog' AND 
    schemaname != 'information_schema';
"""

pd.read_sql(query, con=engine)
```

```python
%time df0.to_sql(name=f'{color}_tripdata_{year}_{month}', con=engine, if_exists='append')
```

```python
df0.info()
```

```python
from time import time
```

```python
while True: 
    try:
        t_start = time()

        df = next(df_iter)

        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)

        df.to_sql(name=f'{color}_tripdata_{year}_{month}', con=engine, if_exists='append')

        t_end = time()

        print('inserted another chunk, took %.3f second' % (t_end - t_start))
    
    except StopIteration:
            print("Finished ingesting data into the postgres database")
            break
```
